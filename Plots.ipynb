```python
steps = np.linspace(0,50000,5000).astype(int)
```

```python
rewards = []
import re
pattern = re.compile("/^[+-]?((\d+(\.\d*)?)|(\.\d+))$/")

for i, line in enumerate(open('ppo_outputs/ppo.txt')):
    rewards.append(float(re.sub("[^\d\.]", "", line)))
rewards = [i / 20 for i in rewards]
plt.plot(steps, rewards)
```
